```python
# this will open the file

file = open('D:/Academics (Masters in DSA)/Sem 2/TA/TA Assignment-1/tweet.txt', encoding='utf8')
read = file.read()
file.seek(0)
read
```

```python
# to obtain the number of lines in file

line = 1
for word in read:
    if word == '\n':
        line+=1
print('Number of lines in file is:', line)

# create a list to store each line as an element of list

array=[]
for i in range(line):
    array.append(file.readline())
array
```

```python
txt = []
for i in range(0,len(array)):
    x=array[i].partition('\t')[2]
    txt.append(x)
print(txt)
```

```python
import re
pattern = "(.*?)\\t"
docID = []
for i in array:
    y = re.search(pattern, i).group(1)
    docID.append(y)
print(docID)
```

```python
text = " ".join(str(x) for x in txt)
text
```

```python
# remove punctuation

punc = '''!()-[]{};:'"\, <>./?@#$%^&*_~'''
for ele in text:   
    if ele in punc:   
        text = text.replace(ele, " ")   
          
text
  
# to maintain uniformity 

text=text.lower()                     
text
```

```python
# Clean data by removing stopwords

import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize 
from nltk.corpus import stopwords 
nltk.download('stopwords') 
  
# this will convert the word into tokens 

for i in range(1):
    text_tokens = word_tokenize(text) 
  
tokens_without_sw = [ 
    word for word in text_tokens if not word in stopwords.words()] 
  
print(tokens_without_sw)
```

```python
#Create an inverted index

dictionary = {} 
  
for i in range(line): 
    check = txt[i].lower() 
    for item in tokens_without_sw: 
  
        if item in check: 
            if item not in dictionary: 
                dictionary[item] = [] 
  
            if item in dictionary: 
                dictionary[item].append(i+1) 
  
dictionary
```

```python
# mapping docID's with natural numbers

b = dict(enumerate(docID, start=1))
print(b)
```

```python
# Natural numbers having same docID's

flipped = {}
  
for key, value in b.items():
    if value not in flipped:
        flipped[value] = [key]
    else:
        flipped[value].append(key)
flipped
```

```python
# Conjunctive query

from functools import reduce
def conj_query(*args):
    q=[] 
    for arg in args:
        p = new_dict.get(arg)
        q.append(p) 
    res = list(reduce(lambda i, j: i & j, (set(n) for n in q)))
    print(res)
```

```python
conj_query('chocolate','strawberry')
```

```python
# Disjunctive query

from itertools import chain
def disj_query(*args):
    q=[] 
    for arg in args:
        p = new_dict.get(arg)
        q.append(p) 
        x = list(chain(*q))
        y = list(dict.fromkeys(x))
    print(y)
```

```python
disj_query('egg', 'cheese')
```

```python
disj_query('chocolate','strawberry')
```

```python
disj_query('egg','cheese','bacon')
```
